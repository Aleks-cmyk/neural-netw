```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import sklearn.datasets as dt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

# Завдання 2

З використанням бібліотеки Sklearn вирішити задачу регресії 

```python
x = np.array([1, 2, 3, 4, 5, 6]).reshape((-1, 1))
y = np.array([10, 45, 80, 115, 150, 185])

print('X : {}'.format(x))
print('Y : {}'.format(y))

model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
print('R^2 - coefficient of determination = {}'.format(r_sq))
print("Linear regression model: y = a + bx")
print('a =', model.intercept_)
print('b =', model.coef_)

test_x = np.array([10, 20, 30]).reshape((-1, 1))
pred_y = model.predict(test_x)
print('Predicted values: {}'.format(pred_y))
```

Згенерувати набір даних за допомогою означеної функції, з використанням якнайбільшої кількості параметрів. Розмірність набору даних прийняти у вигляді (100000х(30-n)), де n - номер за списком. 100000x(30-5)=100000x25=2500000.
Здійснити моделювання згенерованого набору даних та візуалізацію результатів.

```python
color_map = plt.cm.get_cmap('Paired')
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16,7))
plt_ind_list = np.arange(3) + 231

for noise, plt_ind in zip([0, 10, 100], plt_ind_list): 
    x, y = dt.make_regression(n_samples=2500000,
                             n_features=2,
                             noise=noise,
                             random_state=23) 
    
    plt.subplot(plt_ind)
    my_scatter_plot = plt.scatter(x[:,0],
                                  x[:,1],
                                  c=y,
                                  vmin=min(y),
                                  vmax=max(y),
                                  cmap=color_map)
    
    plt.title('noise: ' + str(noise))
    plt.colorbar(my_scatter_plot)
    
fig.subplots_adjust(hspace=0.3, wspace=.3)
plt.suptitle('make_regression()')
plt.show()
```

Надати значення метрик якості, приклад прогнозування

```python
x, y = dt.make_regression(n_targets=1)
y = y.astype(np.uint8)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=42)

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_train_pred = linear_model.predict(x_train)
y_test_pred = linear_model.predict(x_test)

print(f'Mean squared error for trained set: {mean_squared_error(y_train, y_train_pred)}')
print(f'Mean squared error for tested set: {mean_squared_error(y_test, y_test_pred)}')
```
