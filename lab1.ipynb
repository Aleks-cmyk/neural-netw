## створити набір даних лінійної функції з урахуванням помилок вимірювань

```python
x = np.arange(0, 10, 0.5)
y = 7 * x + 0.4
```

## побудувати графік функції та згенерованих даних

```python
plt.figure(figsize=(10,5)) 
 
plt.scatter(x,y, color="blue") 
 
plt.title("Graphic") 
plt.xlabel("X") 
plt.ylabel("Y")

plt.figure(figsize=(10,5)) 
 
plt.plot(x,y, color="blue") 
 
plt.title("Graphic") 
plt.xlabel("X") 
plt.ylabel("Y")
```

## обчислити похибку з використанням метрик MAE, MSE

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("MAE = ", mean_absolute_error(x, y))
print("MSE = ", mean_squared_error(x, y))
```

## Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import csv

FILENAME = "data.csv"

with open(FILENAME, "w", newline="") as file:
  writer = csv.writer(file)
  writer.writerow(["Name", "Value"])
  writer.writerows([["X", x], ["Y", y], ["MAE", mean_absolute_error(x, y)], ["MSE", mean_squared_error(x, y)]])
```
